# GCP Dataset EDA
Run on Kaggle (or anywhere the data is present). Confirms class counts, image-dimension variation, missing labels, and group sizes.

```python
import json, cv2, numpy as np
from pathlib import Path
from collections import Counter
import matplotlib.pyplot as plt

# Adjust to your Kaggle paths:
ROOT = Path('data/train_dataset')
LABELS = json.load(open('data/gcp_marks.json'))
print('entries:', len(LABELS))
```

```python
# Missing verified_shape audit
missing = [k for k, v in LABELS.items() if 'verified_shape' not in v]
print('missing verified_shape:', len(missing))
print(missing[:5])
print(Counter(v.get('verified_shape', 'MISSING') for v in LABELS.values()))
```

```python
# Class distribution bar
counts = Counter(v.get('verified_shape', 'MISSING') for v in LABELS.values())
plt.bar(counts.keys(), counts.values()); plt.title('Class distribution'); plt.ylabel('count')
plt.savefig('outputs/class_dist.png', bbox_inches='tight', dpi=100); plt.show()
print(counts)
```

```python
# Image-dimension variation (sample 150)
dims = []
for k in list(LABELS)[:150]:
    p = ROOT / k
    if p.exists():
        im = cv2.imread(str(p))
        if im is not None: dims.append(im.shape[:2])
print('unique (H,W):', Counter(dims))
```

```python
# Group sizes (near-duplicate frames per GCP folder)
groups = Counter(str(Path(k).parent) for k in LABELS)
print('num groups:', len(groups))
print('median frames/group:', np.median(list(groups.values())))
print('max frames/group:', max(groups.values()))
```

```python
# Coordinate spread (confirms NOT 2048x1365)
xs = [v['mark']['x'] for v in LABELS.values()]; ys = [v['mark']['y'] for v in LABELS.values()]
print('x range', min(xs), max(xs)); print('y range', min(ys), max(ys))
```
